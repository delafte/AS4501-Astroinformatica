In the graphic, we have the PDF curve without any kind of fit in a particular distribution model. I use it as a base to compare it with the fitted in version with a normal distribution and see if they are similar.

To graph the function with a normal distribution, we have to calculate the mean and std of the data. For that, I'm going to use bootstrap.

```python
#First, we see the data that this function provides, so that we can compare it later with the result of bootstrap
julyCalama.describe()
```

```python
julyCalama['HR'].mean()
```

```python
julyCalama['HR'].std()
```

```python
#Let's calculate the recommended sample size
N_sample = len(julyCalama)
N_boot = int(np.round(N_sample * np.log(N_sample)**2))
print("The recomended sample size is {}".format(N_boot))
```

```python
#now that we have the recommended sample size, we make a loop the same amount of times 
means =[]
stds=[]
for i in range(N_boot):
    sample = np.random.choice(julyCalama['HR'], N_sample, replace=True)#the sample
    mean = np.mean(sample)#the mean of the sample
    std = np.std(sample)#the std of the sample
    means.append(mean)#we save the results
    stds.append(std)
bootstrap_means = np.array(means)
bootstrap_stds = np.array(stds)
#now we calculate the means of the results
mean1 = np.mean(bootstrap_means)
std2 = np.mean(bootstrap_stds)
print(mean)
print(std)
```

The values obtained are very similar to the ones that the function describe returns, because of that, for the rest of the regions we are going to use that function.

```python
#Now we use the values obtained to make the gaussian
gaussian_rv = scipy.stats.norm(loc=14.79656796769852, scale=8.139374825909957)
x_to_plot = np.linspace(0, 100, 1000)

plt.hist(julyCalama['HR'], density=True)
plt.plot(x_to_plot,gaussian_rv.pdf(x_to_plot),color='purple',label='July 2023, 2022, 2021')
plt.xlabel('HR')
plt.ylabel('pdf(x)')
plt.title('Calama')
plt.legend(loc='upper right')
plt.show()
```
